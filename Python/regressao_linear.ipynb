# Testes com Python - Regressão Linear

Fazendo alguns testes com modelos de regressão.

```python
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm

x = sp.random.normal(size=100)
z = sp.random.normal(size=100)
categoria = sp.random.binomial(n=1, p=.5, size=100)
erro = sp.random.normal(size=100)

y = 3 + 2*x + 0.65*z + 1.5*categoria + erro
y
```

```python
plt.hist(y, color='red')
plt.title('Histograma da variável resposta')
plt.show()
```

```python
# Montando um modelinho de regressão
dic = {"y":y, "x":x, "categoria":categoria}
dados = pd.DataFrame(data=dic)
dados
```

```python
reg = sm.ols(formula='y~x+categoria', data=dados).fit()
reg.summary()
```
